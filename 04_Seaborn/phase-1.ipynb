- to see the statistical relation between two or more variables,

````python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
````

````python
tips = sns.load_dataset('tips')
tips
````

````python
# scatter plot --> axes level function
# scatter plot for total_bill vs tip
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', style='time', size='size')
````

````python
# scatter plot with figure level function

# relplot --> figure level function --> square shape
# `hue` will add legend in the graph from which we can make distinction easily
# using `style` we can add one more layer of info in the graph
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', hue='sex', style='time', size='size')
````

#### lineplot

````python
# lineplot

# this loads the built-in `gapminder` dataset that comes with Plotly Express
gap = px.data.gapminder()

# filter data for country `India`
temp_df = gap[gap['country'] == 'India']
temp_df
````

````python
# lineplot --> axes level function
sns.lineplot(data=temp_df, x='year', y='lifeExp')
````

````python
# using relplot --> figure level function
sns.relplot(data=temp_df, x='year', y='lifeExp', kind='line')
````

````python
# hue --> style
temp_df = gap[gap['country'].isin(['India', 'Pakistan', 'China'])]
temp_df
````

````python
# line plot side by side (figure level function)
sns.relplot(kind='line',data=temp_df, x='year', y='lifeExp', hue='country')
````

````python
# line plot using axes level function
sns.lineplot(data=temp_df, x='year', y='lifeExp', hue='country')
````

````python
# example - 2
temp_df = gap[gap['country'].isin(['India', 'Brazil', 'Germany'])]
temp_df
````

````python
# example 2 - line plot side by side (figure level function)
sns.relplot(kind='line',data=temp_df, x='year', y='lifeExp', hue='country', style='continent', size='continent')
````

#### facet plot

````python
# facet plot - this will only work with figure level function --> work with relplot
# but it will not work scatterplot and lineplot
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', col='sex', row='day')
````

````python
# facet plot using line plot
sns.relplot(data=tips, x='total_bill', y='tip', kind='line', col='sex', row='day')
````

````python
# facet plot won't work with axis level function scatterplot
# it will throw error if we use
sns.scatterplot(data=tips, x='total_bill', y='tip', col='sex', row='day')
````

````python
# col wrap
sns.relplot(data=gap, x='lifeExp', y='gdpPercap', kind='scatter', col='year', col_wrap=4)
````

### Distribution plot
- used for univariate analysis
- used to find out the distribution
- Range of the observation
- Central Tendency
- is the data bimodal? (multiple peaks)
- Are these outliers?

Plots unders distribution plot
- histplot
- kdplot
- rugplot

````python
# figure level --> displot
# axes level --> histplot , kdeplot , rugplot
````

````python
# using univariate histogram (using tips data)
sns.histplot(data=tips, x='total_bill')
````

````python
# using displot
sns.displot(data=tips, x='total_bill', kind='hist')
````

````python
# bins parameter
sns.displot(data=tips, x='total_bill', kind='hist', bins=20)
````

````python
# It's also possible to visualize the distribution of a categorical variable using the logic of a histogram.
# Discrete bins are automatically set for categorial variable.

# countplot
sns.displot(data=tips, x='day', kind='hist')
````

````python
# hue parameter
sns.displot(data=tips, x='tip', kind='hist', hue='sex')
````

````python
# element --> step (for better visualization)
sns.displot(data=tips, x='tip', kind='hist', hue='sex', element='step')
````

````python
# load titanic dataset from seaborn
titanic = sns.load_dataset('titanic')
titanic
````

````python
sns.displot(data=titanic, x='age', kind='hist', element='step', hue='sex')
````

````python
# faceting using col and row --> this function will not work on histplot function as hisplot is axes level function
sns.displot(data=tips, x='tip', kind='hist', col='sex', element='step')
````

**Seaborn** is a Python's **data visualization** library based on **matplotlib**. It provides a high-level interface for creating attractive and informative statistical graphics.

### 🔹 Why use Seaborn?

- provide a layer of abstraction hence simpler to use
- better aesthetics
- more graphs included

* ✅ **Simplifies complex plots** like heatmaps, violin plots, or pair plots with minimal code.
* ✅ **Beautiful default styles** and themes.
* ✅ **Built-in support for Pandas DataFrames**.
* ✅ **Statistical visualizations** (e.g., regression lines, distribution plots) are easier to make.
* ✅ Handles categorical variables and numeric variables well.

### Summary:

**Seaborn** is ideal when we want **quick, beautiful, and insightful visualizations**—especially with **statistical data**.

---

### 🔶 **Two Types of Functions in Seaborn**

Seaborn provides **two levels of plotting functions**:

#### 1. **Figure-level functions**

* These control the **entire figure**.
* Can create **multiple subplots** automatically.
* Examples: `sns.relplot()`, `sns.catplot()`, `sns.displot()`

#### 2. **Axes-level functions**

* These plot directly on a **given `matplotlib.axes`** object.
* Used for more **custom control**.
* Examples: `sns.scatterplot()`, `sns.lineplot()`, `sns.boxplot()`

🟩 **Both types produce similar plots**, but differ in flexibility and layout control.

---

### 🔷 **Matplotlib Object Hierarchy Recap**

Understanding the plotting structure helps:

* `Figure`: The entire canvas.
* `Axes`: The area where the plot is drawn (can be multiple per figure).
* Inside Axes: titles, labels, lines, markers, ticks, etc.

This hierarchy helps to understand subplotting and layout control in Seaborn.

---

### 🔶 **Main Classifications of Seaborn Plots**

Seaborn plots are categorized into **six major types**:

#### 1. 🟦 **Relational Plots** (Relationship between two variables)
- to see the statistical relation between two or more variables.
- Bivariate Analysis 

* `scatterplot()` → Scatter graph
* `lineplot()` → Line graph (often time-based)

#### 2. 🟧 **Distribution Plots** (Single variable distribution)

* `histplot()` → Histogram
* `kdeplot()` → Kernel Density Estimation
* `rugplot()` → Small ticks on axis

#### 3. 🟨 **Categorical Plots** (Categorical vs numeric)

* `barplot()`, `countplot()`
* `boxplot()`, `violinplot()`, `stripplot()`, `swarmplot()`

#### 4. 🟩 **Matrix Plots** (Gridded data)

* `heatmap()` → Color-coded matrix (great for correlations)
* `clustermap()` → Hierarchical clustering of matrix

#### 5. 🟥 **Regression Plots** (Linear relationships + regression lines)

* `regplot()` → Scatter + regression line
* `lmplot()` → Figure-level version

#### 6. 🟪 **Multi-Plot Grids**

* `jointplot()` → Combines scatter + histograms
* `pairplot()` → All variable pairs in dataset

### 🔷 **Seaborn Relational Plots Summary**

**Relational plots** in Seaborn help visualize the **statistical relationship between two variables** (usually numeric). These are perfect for **bivariate analysis**, such as understanding the relationship between experience and salary.

Seaborn provides **two types of relational plots**:

1. `scatterplot()` – Axis-level function (direct plotting)
2. `relplot(kind="scatter")` – Figure-level function (more flexible layout)

### ✅ **Key Concepts & Important Pointers**

### 🔹 Basic Setup:

* Import required libraries:

  ```python
  import seaborn as sns
  import matplotlib.pyplot as plt
  import pandas as pd
  import numpy as np
  ```

* Load example dataset:

  ```python
  tips = sns.load_dataset("tips")
  ```

---

#### 🔹 1. **Scatter Plot** – Axis-level

```python
sns.scatterplot(data=tips, x="total_bill", y="tip")
```

* Directly plots individual data points.
* Works well with numerical variables.

---

#### 🔹 2. **Relplot with kind='scatter'** – Figure-level

```python
sns.relplot(data=tips, x="total_bill", y="tip", kind="scatter")
```

* Allows **multiple subplots**.
* Better for **multi-faceted or grouped data**.
* Default figure shape is more **square** compared to scatterplot.

---

#### 🆚 **Axis-level vs. Figure-level Functions**

| Feature                 | Axis-level (`scatterplot`) | Figure-level (`relplot`)  |
| ----------------------- | -------------------------- | ------------------------- |
| Control individual axes | ✅                          | ❌ (controlled internally) |
| Multiple subplots       | ❌                          | ✅                         |
| Layout flexibility      | ❌                          | ✅                         |
| Simpler for small tasks | ✅                          | ❌                         |
| Recommended by Seaborn  | ❌ (less scalable)          | ✅                         |

---

#### 🔹 3. **Enhancing Scatter Plots** with Parameters

You can use extra parameters to **encode additional information**:

##### ✅ `hue`

* Adds color based on a categorical column (e.g., `sex`)

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex")
```

##### ✅ `style`

* Changes marker shape based on a categorical column (e.g., `time`)

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", style="time")
```

##### ✅ `size`

* Changes marker size based on a numeric column (e.g., `size`)

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", size="size")
```

---

#### 🔹 4. **Combining Parameters**

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex", style="time", size="size")
```

* Makes the plot multi-dimensional and more insightful.

---

#### 🔹 5. **General Tips**

* Seaborn recommends using **figure-level functions** (like `relplot`) for flexibility and future scalability.
* `relplot()` can be reused for both **scatter** and **line** plots by changing the `kind`:

  ```python
  sns.relplot(data=tips, x="total_bill", y="tip", kind="line")
  ```
* `relplot` = one function, many plot types via `kind=...` → clean and consistent codebase.

#### ✅ **1. Line Plot (Seaborn)**

##### 🔹 What is a Line Plot?

* A **line plot** is used to show **continuity** in data over a continuous variable (usually **time**).
* Ideal for **time series data** (years, months, days).
* Shows trend, progression, or pattern.

##### 🔹 Basic Syntax (Access Level Function):

```python
sns.lineplot(x='experience', y='salary', data=df)
```

##### 🔹 Use Case:

```python
sns.lineplot(x='year', y='lifeExp', data=india_df)
```

---

#### ✅ **2. `relplot()` – Figure-Level Function**

##### 🔹 What is it?

* A **figure-level** function in Seaborn.
* Used for relational plots like **line** and **scatter**.
* Can handle **facets (multiple subplots)** using rows/columns.

##### 🔹 Syntax:

```python
sns.relplot(kind='line', x='year', y='lifeExp', data=df)
```

##### 🔹 Benefits:

* More control over full **figure layout**.
* Supports **faceting** (rows and columns).
* Legends and styling are handled **outside** the plot for better clarity.

---

#### ✅ **3. Filtering Data for Specific Countries**

```python
df[df['country'].isin(['India', 'Pakistan', 'China'])]
```

* Allows comparing multiple countries’ trends side-by-side.

---

#### ✅ **4. Adding More Information to Line Plots**

##### 🔹 `hue`: Differentiate by category (e.g., country)

##### 🔹 `style`: Different line styles (e.g., dashed, dotted based on continent)

##### 🔹 `size`: Controls **thickness** of line

```python
sns.relplot(kind='line', x='year', y='lifeExp', hue='country', style='continent', size='continent', data=df)
```

---

#### ✅ **5. Access-Level vs. Figure-Level Functions**

| Feature            | Access-Level (`lineplot()`, `scatterplot()`) | Figure-Level (`relplot()`, `catplot()`) |
| ------------------ | -------------------------------------------- | --------------------------------------- |
| Scope              | One Axes                                     | Full Figure (Multiple subplots)         |
| Legend             | Inside                                       | Outside                                 |
| Facets             | ❌ Not Supported                              | ✅ Supported                             |
| `col`, `row`       | ❌                                            | ✅                                       |
| More Customization | ❌                                            | ✅                                       |

---

#### ✅ **6. Facet Plot (Faceting using `relplot`)**

##### 🔹 Use Case:

Split the same plot across different **categories**.

```python
sns.relplot(kind='scatter', x='total_bill', y='tip', data=tips, col='sex')
```

* You can also facet on **row**, **col**, or both:

```python
sns.relplot(kind='scatter', x='total_bill', y='tip', data=tips, row='sex', col='smoker')
```

---

#### ✅ **7. `col_wrap` Parameter**

* Controls how many subplots per row.
* Used when too many categories are creating a wide plot.

```python
sns.relplot(kind='line', x='gdpPercap', y='lifeExp', col='year', col_wrap=4, data=gapminder)
```

---

#### ✅ **8. Important Concepts Recap**

* **Line plot** = Trend over time.
* **`relplot()`** = Figure-level, supports `row`, `col`, `hue`, `style`, `size`.
* **`lineplot()`** = Access-level, no faceting.
* **Facet plots** = Subdivide plots for better readability.
* **`hue`** = Color separation by category.
* **`style`** = Line type for extra categorical distinction.
* **`size`** = Line thickness (in line plot) or dot size (in scatter).
* **Use `col_wrap`** to manage wide subplot arrangements.
* Use **`df[df['col'] == value]`** or **`isin()`** for filtering.

---

#### 📌 Tips to Remember

* Use `relplot()` for **complex layouts** (facet/grid plots).
* Use `lineplot()` for **simple trends**.
* Time on **x-axis** → Prefer line plots.
* Legends in `relplot()` appear **outside** for clarity.
* Only **figure-level** functions support `row`, `col`, `col_wrap`.

### 🔷 **What are Distribution Plots?**

Distribution plots help understand **how data is spread** or distributed across values.
They're essential for **univariate analysis**—analyzing one variable at a time.

---

### 📊 **Types of Distribution Plots in Seaborn**

### 1. **Histograms (`histplot()` or `displot(kind="hist")`)**

* Breaks data into **"bins" (intervals)** and shows **count** of data points in each bin.
* Used for **numerical variables**.
* Helps identify:

  * **Range of data**
  * **Central tendency** (mean/median)
  * **Outliers**
  * **Data skewness**

#### 🧪 Example:

```python
sns.histplot(data=tips, x='total_bill')
```

#### 📝 Important Parameters:

* `bins`: Control number of bins (intervals).
* `hue`: Separate histograms for different categories.
* `element`: Show `bars`, `step`, or `poly` (outline).
* `multiple`: Control overlay or stacking of distributions.

---

#### 2. **KDE Plot (`kdeplot()` or `displot(kind="kde")`)**

* **Kernel Density Estimate**: A smooth curve that estimates the distribution.
* Better for visualizing **density shape** compared to histograms.
* Can be used **with or without histograms**.

---

#### 3. **Rug Plot (`rugplot()`)**

* Adds **small vertical ticks** on x-axis for each data point.
* Often used along with KDE to show **actual observations**.

---

### 🧠 **Concepts and Tips Explained**

#### ✅ Why Use Distribution Plots:

* To see how your data is **distributed or spread**.
* Helps in understanding:

  * **Minimum and maximum values**
  * **Where most data points lie**
  * **Whether the data is skewed or symmetrical**
  * **Presence of outliers**

#### ✅ Common Use Case:

* Analyzing **salary distribution**, **customer visits**, etc.

---

### 🔧 **Two Plotting APIs in Seaborn**

| Type             | Function              | Description                                        |
| ---------------- | --------------------- | -------------------------------------------------- |
| **Axes-level**   | `histplot`, `kdeplot` | Direct plotting on given `Axes`.                   |
| **Figure-level** | `displot`             | Creates full figure, supports `col`, `row`, `hue`. |

---

### 📋 **Key Parameters Recap**

| Parameter     | Usage                                                               |
| ------------- | ------------------------------------------------------------------- |
| `x`           | Column to plot on x-axis                                            |
| `data`        | DataFrame containing the data                                       |
| `bins`        | Number of bins for histogram                                        |
| `hue`         | Grouping by a categorical variable (e.g. 'sex')                     |
| `element`     | Show bars (`'bars'`), lines (`'step'`), or outlines (`'poly'`)      |
| `multiple`    | Overlay (`'layer'`), stack (`'stack'`), or dodge (`'dodge'`) groups |
| `col` / `row` | Create **faceted plots** for comparisons across categories          |

---

### 🧪 **Extra Insights**

* You can **use histograms with categorical data** (like days or gender), but it behaves like a **count plot**.
* For **comparison between categories** (e.g. male vs female tips), use `hue`.
* Use `element="step"` to make overlapping histograms clearer.
* You can **facet plots** using `col` or `row` in `displot` for side-by-side comparisons.

---

### 🔚 Summary:

Seaborn’s distribution plots are powerful for:

* Understanding **data shape and spread**
* Visualizing **central values and outliers**
* Comparing **distributions across categories**

#### Use:

* `histplot()` for quick histograms.
* `displot()` for full figure-level customizability.
* `kdeplot()` for smooth density estimates.
* `rugplot()` for raw value tick marks.
